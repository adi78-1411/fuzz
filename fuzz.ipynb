Mileage based on Driving and Engine Condition.


Rules:

1. If the Fule Quality is bad OR the Engine condition is poor, then the Mileage will be low
2. If the Fule Quality is acceptable, then the Mileage will be medium
3. If the Fule Quality is great OR the Engine Condition is amazing, then the Mileage will be high.

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
#   * Fule and Enfine Condition on subjective ranges [0, 5]
#   * Mileage has a range of [0, 100] in units of percentage points
x_Fule = np.arange(0, 6, 1)
x_Eng = np.arange(0, 6, 1)
x_Mileage  = np.arange(0, 101, 1)
```

```python
# Generate fuzzy membership functions
# Fule quality
Fule_lo = fuzz.trimf(x_Fule, abc=[0, 0, 2])
Fule_md = fuzz.trimf(x_Fule, abc=[0, 3, 5])
Fule_hi = fuzz.trimf(x_Fule, abc=[2, 5, 5])
# Engine Condition
Eng_lo = fuzz.trimf(x_Eng, abc=[0, 0, 2])
Eng_md = fuzz.trimf(x_Eng, abc=[0, 3, 5])
Eng_hi = fuzz.trimf(x_Eng, abc=[2, 5, 5])
# Mileage 
Mileage_lo = fuzz.trimf(x_Mileage, abc=[0, 0, 50])
Mileage_md = fuzz.trimf(x_Mileage, abc=[0, 50, 100])
Mileage_hi = fuzz.trimf(x_Mileage, abc=[50, 100, 100])
```

```python
print('x_Mileage  ',x_Mileage  )
print('Mileage_lo', Mileage_lo)
print('Mileage_md',Mileage_md)
print('Mileage_hi',Mileage_hi)
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))

ax0.plot(x_Fule, Fule_lo, 'b', linewidth=1.5, label='Bad')
ax0.plot(x_Fule, Fule_md, 'g', linewidth=1.5, label='Decent')
ax0.plot(x_Fule, Fule_hi, 'r', linewidth=1.5, label='Great')
ax0.set_title('Fule quality')
ax0.legend()

ax1.plot(x_Eng, Eng_lo, 'b', linewidth=1.5, label='Poor')
ax1.plot(x_Eng, Eng_md, 'g', linewidth=1.5, label='Acceptable')
ax1.plot(x_Eng, Eng_hi, 'r', linewidth=1.5, label='Amazing')
ax1.set_title('Engine Condition')
ax1.legend()

ax2.plot(x_Mileage, Mileage_lo, 'b', linewidth=1.5, label='Low')
ax2.plot(x_Mileage, Mileage_md, 'g', linewidth=1.5, label='Medium')
ax2.plot(x_Mileage, Mileage_hi, 'r', linewidth=1.5, label='High')
ax2.set_title('Mileage')
ax2.legend()

# Turn off top/right axes
for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

What would the tip be in the following circumstance:

Fule quality was 3.9
Engine Condition was 1.9

```python
Fule_level_lo = fuzz.interp_membership(x_Fule, Fule_lo, 3.5)
Fule_level_md = fuzz.interp_membership(x_Fule, Fule_md, 3.5)
Fule_level_hi = fuzz.interp_membership(x_Fule, Fule_hi, 3.5)
print(Fule_level_lo, Fule_level_md, Fule_level_hi)
```

```python
Eng_level_lo = fuzz.interp_membership(x_Eng, Eng_lo, 1.9)
Eng_level_md = fuzz.interp_membership(x_Eng, Eng_md, 1.9)
Eng_level_hi = fuzz.interp_membership(x_Eng, Eng_hi, 1.9)
print(Eng_level_lo, Eng_level_md, Eng_level_hi)
```

1. If the Fule Quality is bad OR the Engine condition is poor, then the Mileage will be low

```python
import math 

active_rule1 = np.fmax(Fule_level_lo, Eng_level_lo) # fuzzy-max = Boolean OR
print('active_rule1', active_rule1)
# Now we apply this by clipping the top off the corresponding output
# membership function with `np.fmin`
print(Mileage_lo.round(2))
Mileage_activation_lo = np.fmin(active_rule1, Mileage_lo)  # fuzzy-min = Boolean AND
print(Mileage_activation_lo)
```

2. If the Fule Quality is acceptable, then the Mileage will be medium


```python
# For rule 2 we connect Fule Quality to medium Mileage
Mileage_activation_md = np.fmin(Fule_level_lo, Mileage_lo) # fuzzy-min = Boolean AND
print(Mileage_md.round(1))
print(Mileage_activation_lo)
```

3. If the Fule Quality is great OR the Engine Condition is amazing, then the Mileage will be high.


```python
active_rule3 = np.fmax(Fule_level_hi, Eng_level_hi) # fuzzy-max = Boolean OR
print('active_rule3', active_rule3)
# Now we apply this by clipping the top off the corresponding output
# membership function with `np.fmin`
print(Mileage_hi)
Mileage_activation_hi = np.fmin(active_rule3, Mileage_hi)  # fuzzy-min = Boolean AND
print(Mileage_activation_hi)
```

```python
Mileage_ = np.zeros_like(x_Mileage)
# Visualize this
fig, ax0 = plt.subplots(figsize=(8, 3))
ax0.fill_between(x=x_Mileage, y1=Mileage_, y2=Mileage_activation_lo, facecolor='b', alpha=0.5)
ax0.plot(x_Mileage, Mileage_lo, 'b', linewidth=0.5, linestyle='--', )
ax0.fill_between(x_Mileage, Mileage_, Mileage_activation_lo, facecolor='g', alpha=0.5)
ax0.plot(x_Mileage, Mileage_md, 'g', linewidth=0.5, linestyle='--')
ax0.fill_between(x_Mileage, Mileage_, Mileage_activation_hi, facecolor='r', alpha=0.5)
ax0.plot(x_Mileage, Mileage_hi, 'r', linewidth=0.5, linestyle='--')
ax0.set_title('Output membership activity: Rule3+Rule2+Rule3')

# Turn off top/right axes
for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

- Aggregate all three output membership functions together


```python
aggregated = np.fmax(Mileage_activation_lo, np.fmax(Mileage_activation_md, Mileage_activation_hi))
print(aggregated)
```

Defuzzification
Calculate defuzzified result using the All method

```python
Mileage = fuzz.defuzz(x_Mileage, aggregated, 'centroid')
print(Mileage)
Mileage_s = fuzz.defuzz(x_Mileage, aggregated, 'som')
print(Mileage_s)
Mileage_m = fuzz.defuzz(x_Mileage, aggregated, 'mom')
print(Mileage_m)
Mileage_l = fuzz.defuzz(x_Mileage, aggregated, 'lom')
print(Mileage_l)
Mileage_b= fuzz.defuzz(x_Mileage,aggregated,"Bisector")
print(Mileage_b)
```

```python
Mileage_activation = fuzz.interp_membership(x_Mileage, aggregated, Mileage)  # for plot
Mileage_activation
```

API

```python
from skfuzzy import control as ctrl
Fule = ctrl.Antecedent(np.arange(0, 6, 1), 'Fule')
Engine_Quality = ctrl.Antecedent(np.arange(0, 6, 1), 'Engine_Quality')
Mileage = ctrl.Consequent(np.arange(0, 101, 1), 'Mileage')
```

Membership Functions


```python
Fule.automf(5)
Fule.view()
Engine_Quality.automf(5)
Engine_Quality.view()
Mileage.automf(5)
Mileage.view()
```

Rules

```python
rule1 = ctrl.Rule(Fule['poor'] | Engine_Quality['poor'], Mileage['poor'])
rule2 = ctrl.Rule(Fule['mediocre'], Mileage['mediocre'])
rule3 = ctrl.Rule(Fule['average'], Mileage['average'])
rule4 = ctrl.Rule(Fule['decent'], Mileage['decent'])
rule5 = ctrl.Rule(Fule['good'] | Engine_Quality['good'], Mileage['good'])
```

```python
Mileage_ctrl = ctrl.ControlSystem([rule1, rule2, rule3,rule4,rule5])
Mileage_clc = ctrl.ControlSystemSimulation(Mileage_ctrl)
Mileage_clc.input['Fule'] = 3.9
Mileage_clc.input['Engine_Quality'] = 1.9

# Crunch the numbers
Mileage_clc.compute()
print (Mileage_clc.output['Mileage'])
Mileage.view(sim=Mileage_clc)
```
